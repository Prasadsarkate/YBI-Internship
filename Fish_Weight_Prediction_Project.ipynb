```python
# Step 1: Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

# Step 2: Import Data
fish = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Fish.csv')

# Step 3: Data Overview (Optional - just to check data)
print("Data Head:")
print(fish.head())
print("\nData Info:")
print(fish.info())
print("\nData Description:")
print(fish.describe())

# Step 4: Define Target (y) and Features (X)
# Hamein 'Weight' predict karna hai based on other features
fish.columns
y = fish['Weight']
X = fish[['Category', 'Height', 'Width', 'Length1', 'Length2', 'Length3']]

# Step 5: Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=2529)

# Step 6: Model Selection
model = LinearRegression()

# Step 7: Train Model
model.fit(X_train, y_train)

# Step 8: Model Prediction
y_pred = model.predict(X_test)

# Step 9: Model Evaluation
print("\nModel Accuracy (MAE):", mean_absolute_error(y_test, y_pred))
print("Model Accuracy (MAPE):", mean_absolute_percentage_error(y_test, y_pred))

print("\nSuccess! Project run complete.")
```
